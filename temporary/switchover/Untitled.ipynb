```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
gainData = pd.read_csv('autogain2.csv')
timeData = gainData[" Time [s]"]
muxData = gainData["Channel 1-Analog"]
lowGain = gainData["Channel 2-Analog"]
highGain = gainData["Channel 3-Analog"]
```

```python
i_start = np.where(timeData == 3)[0][0]
i_end = np.where(timeData == 10)[0][0]

plt.subplot(3,1,1)
plt.plot(timeData[i_start:i_end], muxData[i_start:i_end])
plt.xticks([])
plt.grid()
plt.subplot(3,1,2)
plt.plot(timeData[i_start:i_end], lowGain[i_start:i_end])
plt.xticks([])
plt.grid()
plt.subplot(3,1,3)
plt.plot(timeData[i_start:i_end], highGain[i_start:i_end])
plt.xticks(fontsize = 12)
plt.grid()
plt.xlabel('Time [s]', fontsize = 15)
plt.tight_layout()
plt.savefig('autogain.pdf')
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, sharey=True, figsize=(6,4))

axes[0].plot(timeData[i_start:i_end], muxData[i_start:i_end], color='blue')
axes[0].set_xlabel('')
axes[0].set_title('Switchover Mechanism', fontsize = 13)
axes[0].grid(True)

axes[1].plot(timeData[i_start:i_end], lowGain[i_start:i_end], color='blue')
axes[1].set_title('Low Gain', fontsize = 13)
axes[1].grid(True)

axes[2].plot(timeData[i_start:i_end], highGain[i_start:i_end], color='red')
axes[2].set_title('High Gain', fontsize = 13)
axes[2].grid(True)

fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
fig.tight_layout()

plt.xlabel("Time [s]", fontsize = 13)
plt.ylabel("Voltage [V]", fontsize = 13)
plt.savefig('autogain.pdf')
```
