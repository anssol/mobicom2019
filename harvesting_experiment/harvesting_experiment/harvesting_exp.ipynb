```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# Light levels (lx)
lightLevels = np.array([500, 800, 1500, 2000, 2400])

# Cold start (s)
cs_1500lx = np.array([506.83])
cs_2000lx = np.array([386.7])
cs_2400lx = np.array([296.4])
cs = np.array([cs_1500lx, cs_2000lx, cs_2400lx])

# Charge time (s) (above cold start)
ct_500lx = np.array([217.29, 200.87, 203.5])
ct_800lx = np.array([124.5, 111.14, 110.85])
ct_1500lx = np.array([46.67, 43.74, 43.4, 42.07])
ct_2000lx = np.array([28.3, 27.75, 27])
ct_2400lx = np.array([27.93, 25.21, 24.67, 23.8])
ct = np.array([np.mean(ct_500lx), np.mean(ct_800lx), np.mean(ct_1500lx), np.mean(ct_2000lx), np.mean(ct_2400lx)])
ct_std = np.array([np.std(ct_500lx), np.std(ct_800lx), np.std(ct_1500lx), np.std(ct_2000lx), np.std(ct_2400lx)])

# Discharge times(s)
dt_500lx = np.array([22.692, 18, 21.6, 18.8])
dt_800lx = np.array([32, 29.571, 29.921, 29.235])
dt_1500lx = np.array([58.296, 54.69, 50.57, 51.613])
dt_2000lx = np.array([80.18, 82.914, 83.057, 80.581])
dt_2400lx = np.array([152, 152.793, 142.2, 156])
dt = np.array([np.mean(dt_500lx), np.mean(dt_800lx), np.mean(dt_1500lx), np.mean(dt_2000lx), np.mean(dt_2400lx)])
dt_std = np.array([np.std(dt_500lx), np.std(dt_800lx), np.std(dt_1500lx), np.std(dt_2000lx), np.std(dt_2400lx)])

# High-bandwidth (weird) amplifier discharge times (s)
dt_500lx_ahb = np.array([14.4, 12.3, 12.68, 11.29, 15.5])
dt_800lx_ahb = np.array([12.1, 15.76, 11.2, 14.3])
dt_1500lx_ahb = np.array([10.35, 9.39, 9.33, 7.72])
dt_2000lx_ahb = np.array([9.7, 8.8, 8.4, 8.84])
dt_2400lx_ahb = np.array([12.55, 12.32, 12.03, 12.0])
dt_ahb = np.array([np.mean(dt_500lx_ahb), np.mean(dt_800lx_ahb), np.mean(dt_1500lx_ahb), np.mean(dt_2000lx_ahb), np.mean(dt_2400lx_ahb)])
dt_std_ahb = np.array([np.std(dt_500lx_ahb), np.std(dt_800lx_ahb), np.std(dt_1500lx_ahb), np.std(dt_2000lx_ahb), np.std(dt_2400lx_ahb)])
```

```python
# Plot charge times as a function of light levels
plt.figure(figsize=(7,3))
plt.plot(lightLevels, ct, marker = 'o', markersize = 7, color = 'blue')
plt.errorbar(lightLevels, ct, ct_std, linestyle = 'None', color = 'blue', capsize = 3, elinewidth = 2, markeredgewidth = 2)
plt.grid()
plt.xlabel('Light level [lx]', fontsize = 15)
plt.ylabel('Charge time [s]', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.tight_layout()
plt.savefig('chargetime.pdf')
plt.show()
```

```python
# Plot charge times as a function of light levels
plt.figure(figsize=(7,3))
plt.plot(lightLevels, dt, marker = 'o', markersize = 7, color = 'red')
plt.errorbar(lightLevels, dt, dt_std, linestyle = 'None', color = 'red', capsize = 3, elinewidth = 2, markeredgewidth = 2)
plt.grid()
plt.xlabel('Light Level [lx]', fontsize = 15)
plt.ylabel('Discharge time [s]', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.tight_layout()
plt.savefig('dischargetime.pdf')
plt.show()
```

```python
# (WEIRD BEHAVIOUR) Plot charge times as a function of light levels
plt.figure(figsize=(7,3))
plt.plot(lightLevels, dt_ahb, marker = 'o', markersize = 7, color = 'red')
plt.errorbar(lightLevels, dt_ahb, dt_std_ahb, linestyle = 'None', color = 'red', capsize = 3, elinewidth = 2, markeredgewidth = 2)
plt.grid()
plt.xlabel('Light Level [lx]', fontsize = 15)
plt.ylabel('Discharge time [s]', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.tight_layout()
plt.savefig('hbw_dischargetime.pdf')
plt.show()
```

```python
# Plot charge times as a function of light levels
plt.figure(figsize=(7,3))
plt.plot(lightLevels[2:], cs, marker = 'o', markersize = 7, color = 'green')
plt.grid()
plt.xlabel('Light level [lx]', fontsize = 15)
plt.ylabel('Cold start [s]', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.tight_layout()
plt.savefig('coldstart.pdf')
plt.show()
```

```python
# Plot charge times as a function of light levels
plt.figure(figsize=(7,3))
plt.plot(lightLevels, ct, label = 'Energy harvesting time', marker = 'o', markersize = 7, linewidth=2, color='#3182bd')
plt.errorbar(lightLevels, ct, ct_std, linestyle = 'None', color = '#3182bd', linewidth=2, capsize = 3, elinewidth = 2, markeredgewidth = 2)
plt.plot(lightLevels, dt, label = 'Operational time of active receiver', linewidth=2, marker='o', color = '#de2d26', markersize = 7)
plt.errorbar(lightLevels, dt, dt_std, linestyle = 'None', color = '#de2d26', linewidth=2, capsize = 3, elinewidth = 2, markeredgewidth = 2)
plt.grid()
plt.xlabel('Light level [lx]', fontsize = 15)
plt.ylabel('Time [s]', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(prop = {'size': 14})
plt.tight_layout()
plt.savefig('harvestingexp.pdf')
plt.show()
```
